# 1- Stages of Larage Langage model

Creating a LLM  = Pretraining + Fine-tuning

## Stage 1 of LLM -- Pretraining

- Training on a large, diverse Dataset

- It was done on the unabled text dataset

## Stage 2 of LLM -- Finetuning

- Refinement by training on narrow datatset, specific to a paticular task or domain
- It was done on the label text dataset

### 2 type of fintuning exist

- Instruction Finetunung

Labeled dataset consist of instruction - answer pairs e.g. Text translation air line customer support

- Fintuning for classification Task

Labed Dataset consist of text & associated labels e.g. Email -> Spam vs Non Spam
